```python
# Membuka file Biodata.txt dan mengambil data Nama, Program Studi, NPM, dan Alamat
with open('/resources/data/Biodata.txt', 'r') as readfile:
    lines = readfile.readlines()

# Baris yang mengandung Nama, Program Studi, NPM, dan Alamat
the_data = []
for line in lines:
    if "Nama" in line or "Program Studi" in line or "NPM" in line or "Alamat" in line:
        the_data.append(line)

# Menyimpan data yang diambil ke dalam file txt yang baru
with open('/resources/data/Biodata_2.txt', 'w') as writefile:
    writefile.writelines(the_data)
```
